```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from pdf2docx import Converter
from docx2pdf import convert

def convert_pdf_to_doc(input_pdf_path, output_doc_path):
    cv = Converter(input_pdf_path)
    cv.convert(output_doc_path, start=0, end=None)
    cv.close()
    messagebox.showinfo("Conversion Complete", "PDF to DOC conversion is done!")

def convert_doc_to_pdf(input_doc_path, output_pdf_path):
    convert(input_doc_path, output_pdf_path)
    messagebox.showinfo("Conversion Complete", "DOC to PDF conversion is done!")

def select_pdf_file():
    file_path = filedialog.askopenfilename(filetypes=[("PDF Files", "*.pdf")])
    pdf_file_entry.delete(0, tk.END)
    pdf_file_entry.insert(0, file_path)

def select_doc_file():
    file_path = filedialog.askopenfilename(filetypes=[("DOC Files", "*.docx")])
    doc_file_entry.delete(0, tk.END)
    doc_file_entry.insert(0, file_path)

def clear_file_entries():
    pdf_file_entry.delete(0, tk.END)
    doc_file_entry.delete(0, tk.END)

def convert_file():
    if conversion_var.get() == "pdf_to_doc":
        input_pdf_path = pdf_file_entry.get()
        output_doc_path = doc_file_entry.get()
        convert_pdf_to_doc(input_pdf_path, output_doc_path)
    elif conversion_var.get() == "doc_to_pdf":
        input_doc_path = doc_file_entry.get()
        output_pdf_path = pdf_file_entry.get()
        convert_doc_to_pdf(input_doc_path, output_pdf_path)

root = tk.Tk()
root.title("Shuvo.py")
root.configure(bg="hotpink")  # Set the background color to hot pink

conversion_var = tk.StringVar()
conversion_var.set("pdf_to_doc")

pdf_label = tk.Label(root, text="PDF File:")
pdf_label.grid(row=0, column=0, padx=10, pady=10)
pdf_file_entry = tk.Entry(root, width=50)
pdf_file_entry.grid(row=0, column=1, padx=10, pady=10)
pdf_select_button = tk.Button(root, text="Select", command=select_pdf_file)
pdf_select_button.grid(row=0, column=2, padx=10, pady=10)

doc_label = tk.Label(root, text="DOC File:")
doc_label.grid(row=1, column=0, padx=10, pady=10)
doc_file_entry = tk.Entry(root, width=50)
doc_file_entry.grid(row=1, column=1, padx=10, pady=10)
doc_select_button = tk.Button(root, text="Select", command=select_doc_file)
doc_select_button.grid(row=1, column=2, padx=10, pady=10)

clear_button = tk.Button(root, text="Clear", command=clear_file_entries)
clear_button.grid(row=1, column=3, padx=10, pady=10)

conversion_label = tk.Label(root, text="Conversion:")
conversion_label.grid(row=2, column=0, padx=10, pady=10)
pdf_to_doc_radio = tk.Radiobutton(root, text="PDF to DOC", variable=conversion_var, value="pdf_to_doc")
pdf_to_doc_radio.grid(row=2, column=1, padx=10, pady=10)
doc_to_pdf_radio = tk.Radiobutton(root, text="DOC to PDF", variable=conversion_var, value="doc_to_pdf")
doc_to_pdf_radio.grid(row=3, column=1, padx=10, pady=10)

convert_button = tk.Button(root, text="Convert", command=convert_file)
convert_button.grid(row=4, column=1, padx=10, pady=10)

root.mainloop()
```
